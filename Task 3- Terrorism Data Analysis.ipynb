# Data Visualization

```python
plt.subplots(figsize=(15,6))
sns.countplot(x='attacktype1',data=df)
plt.xlabel('Attack type')
plt.ylabel('No.of attacks')
```

```python
plt.figure(figsize=(31,20),dpi=400)
sns.barplot(df['provstate'].value_counts()[:15].index,df['provstate'].value_counts()[:15].values)
plt.title("Top 10 areas that are attacked")
plt.xlabel("Area")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(31,20),dpi=400)
sns.countplot(x='iyear',data=df,color='black')
plt.xlabel('year')
plt.ylabel('number of attacks')
plt.title('terrorist attacks every year')
```

```python
df['city'].value_counts().nsmallest(10).plot(kind='bar', figsize=(10,5),color='green')
```

```python
df['provstate'].value_counts().nsmallest(10).plot(kind='bar', figsize=(10,5),color='orange')
```
